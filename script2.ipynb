```python
import translators.server as tss
from bs4 import BeautifulSoup, NavigableString
```

```python
def findString(cntnt, elem):
  for i in range(0, len(cntnt)):
    if type(cntnt[i]) == NavigableString:
      if len(cntnt[i])>1:
        translation=tss.google(cntnt[i], to_language="hi")
        elem.find(string=str(cntnt[i])).replace_with(translation)
    else:
      try:
        findString(cntnt[i].contents, elem)
      except: continue
```

```python
def translatePage(f):
    html = open(f, "r", encoding="utf-8")
    soup = BeautifulSoup(html)
    elems = soup.findAll(["title", "body"])
    for elem in elems:
        findString(elem.contents, elem)
    replacedHtml = open(f, "w", encoding="utf-8")
    replacedHtml.write(str(soup.prettify()))
```

```python
htmlFiles = []
import os
for root, dirs, files in os.walk("..\\classcentral3\\www.classcentral.com"):
    for file in files:
        if file.endswith(".html"):
             htmlFiles.append(os.path.join(root, file))
```

```python
success, fail = [], []
for f in htmlFiles[93:]:
    try:
        translatePage(f)
        success.append(f)
        print("(.\\" + f + ")" + " => succeeded.")
    except: 
        print("(.\\" + f + ")" + " => faild.")
        fail.append(f)
```
