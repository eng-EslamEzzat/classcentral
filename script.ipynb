```python
import translators.server as tss
from bs4 import BeautifulSoup
```

```python
def translatePage(filename):
    html = open(filename, "r", encoding="utf-8")
    soup = BeautifulSoup(html)
    elems = soup.findAll(["p","li","h1","h2","h3","h4","h5","h6","td","title","span","strong","button","a"])
    for elem in elems:
        substrings = elem.find_all(string=True)
        if substrings:
            for st in substrings:
                if st and len(st) > 1:
                    fixed_text = st.replace(st, tss.google(st, to_language="hi"))
                    st.replace_with(fixed_text)
            
    replacedHtml = open(filename, "w", encoding="utf-8")
    replacedHtml.write(str(soup.prettify()))
```

```python
htmlFiles = []
import os
for root, dirs, files in os.walk("..\\classcentral3"):
    for file in files:
        if file.endswith(".html"):
             htmlFiles.append(os.path.join(root, file))
```

```python
success, fail = [], []
for f in htmlFiles:
    try:
        translatePage(f)
        success.append(f)
        print("(.\\" + f + ")" + " => succeeded.")
    except: 
        print("(.\\" + f + ")" + " => faild.")
        fail.append(f)
```
